```python

name_list = []
open_value_list = []
max_list=[]
min_list=[]
close_list=[]
difference_list=[]
difference_rate_list=[]
avg_value_list=[]
deal_list=[]
deal_value_list=[]
version_date_list=[]
for i in range(1,10): 
    print('======获取第{}页信息======'.format(i))
    link_t1_list = []
    url_t1 = 'https://www.sge.com.cn/sjzx/mrhqsj?p={}'.format(i)
    respond_t1 = requests.get(url=url_t1,headers=header)
    respond_t1_page = respond_t1.text
    html_t1 = etree.HTML(respond_t1_page)
    link_t1 = html_t1.xpath(r"//div[@class='articleList border_ea mt30 mb30']//a/@href")
    link_t1_list.append(link_t1)
    for link in link_t1_list[0]:
        url_t2 = 'https://www.sge.com.cn{}'.format(link)
        respond_t2 = requests.get(url=url_t2,headers=header)
        respond_t2_page = respond_t2.text
        html_t2 = etree.HTML(respond_t2_page)
        name = html_t2.xpath(r'//tbody//tr[6]//td[1]/text()')
        name_list.append(name)
        open_value = html_t2.xpath(r'//tbody//tr[6]//td[2]/text()')
        open_value_list.append(open_value)
        max_value = html_t2.xpath(r'//tbody//tr[6]//td[3]/text()')
        max_list.append(max_value)
        min_value = html_t2.xpath(r'//tbody//tr[6]//td[4]/text()')
        min_list.append(min_value)
        close_value = html_t2.xpath(r'//tbody//tr[6]//td[5]/text()')
        close_list.append(close_value)
        difference = html_t2.xpath(r'//tbody//tr[6]//td[6]/text()')
        difference_list.append(difference)
        difference_rate = html_t2.xpath(r'//tbody//tr[6]//td[7]/text()')
        difference_rate_list.append(difference_rate)
        avg_value = html_t2.xpath(r'//tbody//tr[6]//td[8]/text()')
        avg_value_list.append(avg_value)
        deal_counts = html_t2.xpath(r'//tbody//tr[6]//td[9]/text()')
        deal_list.append(deal_counts)
        deal_value = html_t2.xpath(r'//tbody//tr[6]//td[10]/text()')
        deal_value_list.append(deal_value)
        version_date = html_t2.xpath(r"//div[@class='title']//h1/text()")
        version_date_list.append(version_date)
```

```python
columns_name =["version_date","name_list",\
               'open_value_list',\
               'max_list','min_list',\
               'close_list','difference_list',\
               'difference_rate_list',\
               'avg_value_list','deal_list',\
               'deal_value_list']
```

```python
#删除空格、换行符
def delete_space(columns_name):
    columns_name = pd.DataFrame(columns_name)
    columns_name[0] = columns_name[0].str.replace('\r','')
    columns_name[0] = columns_name[0].str.replace('\n','')
    columns_name[0]= columns_name[0].str.replace('\t','')
    return columns_name
```

```python
version_date_list = delete_space(version_date_list)
name_list = delete_space(name_list)
open_value_list = delete_space(open_value_list)
max_list = delete_space(max_list)
min_list = delete_space(min_list)
close_list = delete_space(close_list)
difference_list = delete_space(difference_list)
difference_rate_list = delete_space(difference_rate_list)
avg_value_list = delete_space(avg_value_list)
deal_list = delete_space(deal_list)
deal_value_list = delete_space(deal_value_list)
```

```python
data = pd.concat([version_date_list,name_list,open_value_list,max_list,\
                  min_list,close_list,difference_list,\
                 difference_rate_list,avg_value_list,\
                 deal_list,deal_value_list],axis=1)
```

```python
data.columns = columns_name
```
