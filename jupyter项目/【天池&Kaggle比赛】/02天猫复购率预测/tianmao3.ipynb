```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
import copy
import gc
from collections import Counter

warnings.filterwarnings("ignore")

pd.set_option("display.max_column",100)
pd.set_option("expand_frame_repr",False)

all_data_test=pd.read_csv(r"C:\Users\xiongyuan\Desktop\all_data_test.csv",index_col=False)
```

```python
all_data_test.head()
```

```python
user_col_cnt=all_data_test["action_type_path"]
```

```python
user_col_cnt=user_col_cnt.str.split(" ")
user_col_cnt
```

```python
action_total_0=[]
action_total_1=[]
action_total_2=[]
action_total_3=[]
for col in user_col_cnt:
    action_0=[]
    action_1=[]
    action_2=[]
    action_3=[]
    for col_i in col:
        if col_i =="0":
            action_0.append(col_i)
        elif col_i =="1":
            action_1.append(col_i)
        elif col_i =="2":
            action_2.append(col_i)
        elif col_i =="3":
            action_3.append(col_i)
    action_total_0.append(len(action_0))
    action_total_1.append(len(action_1))
    action_total_2.append(len(action_2))
    action_total_3.append(len(action_3))
```

```python
action_total_0=pd.DataFrame(action_total_0)
action_total_1=pd.DataFrame(action_total_1)
action_total_2=pd.DataFrame(action_total_2)
action_total_3=pd.DataFrame(action_total_3)
```

```python
action_total_0.columns=["type_0"]
action_total_1.columns=["type_1"]
action_total_2.columns=["type_2"]
action_total_3.columns=["type_3"]
```

```python
all_data_test=pd.concat([all_data_test,action_total_0,action_total_1,action_total_2,action_total_3],axis=1)
```
