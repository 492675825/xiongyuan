```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

```python
df=pd.read_csv(r'D:\文件\学习\数据集\HR\data\HR.csv')
#删除缺失值
df=df.dropna()
#删除department中的错误类别sale
df=df[df['department']!='sale']
```

```python
df.info()
```

```python
#离散化
```

```python
dep_lab=df[['department']]
from sklearn.preprocessing import LabelEncoder
scaler_dep=LabelEncoder()
dep_lab=scaler_dep.fit_transform(dep_lab)
```

```python
dep_lab
```

```python
dep_list={index:label for index,label in enumerate(scaler_dep.classes_)}
```

```python
dep_list
```

```python
df['dep_LableEncoder']=dep_lab
```

```python
dep_dummy=pd.get_dummies(df['department'],prefix='dep')
```

```python
df=pd.concat([df,dep_dummy],axis=1)
```

```python
df.head()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2,
                        interaction_only=False,
                        include_bias=False)
ls_np=df.iloc[:,1:3]
```

```python
ls_np2=poly.fit_transform(ls_np)
```

```python
ls_np2
```

```python
ls_np2=pd.DataFrame(ls_np2)
```

```python
ls_np2.columns=['ls_np_{}'.format(i) for i in range(1,6)]
```

```python
ls_np2
```

```python
df=pd.concat([df,ls_np2],axis=1)
```

```python
df.head()
```

```python
np.mean(df['last_evaluation'])
```

```python
from sklearn.preprocessing import Binarizer
bin_=Binarizer(threshold=np.mean(df['last_evaluation']))
scaler_eval=bin_.fit_transform(df[['last_evaluation']])
```

```python
df['bin_eval']=scaler_eval
```

```python
df.head()
```

```python
st_le=df['satisfaction_level']
st_le_=pd.cut(x=st_le,bins=5,labels=range(1,6))
```

```python
df['st_le']=st_le_
```

```python
corpus=[
    'The sky is blue and beautiful',
    'Love this blue and beautiful sky',
    'The quick brown fox jumps over the lazy dog',
    'The brown fox is quick and the blue dog is lazy',
    'The sky is very blue and the sky is very beautiful today',
    'The dog is lazy but the brown fox is quick'
]
labels=['weather','weather','animals','animals','weather','animals']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
from sklearn.feature_extraction.text import CountVectorizer
```

```python
corpus=pd.DataFrame(corpus,columns=['text'])
labels=pd.DataFrame(labels,columns=['label'])
```

```python
word=pd.concat([corpus,labels],axis=1)
```

```python
word
```

```python
vocb=TfidfVectorizer(
    stop_words=ENGLISH_STOP_WORDS,
    min_df=0.,
    max_df=1.,
    max_features=100,
    ngram_range=(1,1))
```

```python
words=vocb.fit_transform(word['text'])
```

```python
words
```

```python
words=words.toarray()
```

```python
words
```

```python
words=np.array(words)
```

```python
words
```

```python
vocab_list=vocb.get_feature_names()
```

```python
vocab_list
```

```python
words=pd.DataFrame(np.round(words,2),columns=vocab_list)
```

```python
words
```
