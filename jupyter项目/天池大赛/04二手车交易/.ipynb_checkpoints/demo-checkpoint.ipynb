```python
#构造统计字段数据个数的函数
def add_count_new_feature(df,col_name,new_feature_name):
    df_col=df[[col_name]]
    df_col[new_feature_name]=1
    df_col=df_col.groupby(col_name).agg('sum').reset_index()
    df=df.merge(df_col,on=col_name,how='left')
    return df
```

```python
#调用函数
train_data=add_count_new_feature(train_data,'name','name_count')
train_data.head()
```

```python
#生成brand的计数列
train_data=add_count_new_feature(train_data,'brand','brand_count')
train_data.head()
```

```python
#生成字段bodyType的计数列
train_data=add_count_new_feature(train_data,'bodyType','bodeType_count')
train_data.head()
```

```python
#生成字段fuelType字段计数列
train_data=add_count_new_feature(train_data,'fuelType','fuelType_count')
train_data.head()
```

```python
#生成地区编码regionCode的字段的计数列
train_data=add_count_new_feature(train_data,'regionCode','fuelType_count')
train_data.head()
```

```python
#删去无用的特征SaleID
train_data=train_data.drop('SaleID',axis=1)
train_data.head()
```

```python
#标准化
from sklearn.preprocessing import StandardScaler
data=train_data.drop('price',axis=1)
label=train_data['price']
data=StandardScaler().fit_transform(data)
```

```python
data=pd.DataFrame(data)
data.columns=train_data.drop('price',axis=1).columns
```

```python
data.head()
```

```python
#使用Kfolder切分数据集
from sklearn.model_selection import KFold
kf=KFold(n_splits=5)
for train_index,test_index in kf.split(data):
    x_train=data.values[train_index]
    x_test=data.values[test_index]
    y_train=label.values[train_index]
    y_test=label.values[test_index]
```

```python
#决策树算法
from sklearn.tree import DecisionTreeRegressor
model_dtr=DecisionTreeRegressor()
model_dtr.fit(x_train,y_train)
model_dtr.score(x_test,y_test)
```

```python
#随机森林
from sklearn.ensemble import RandomForestRegressor
model_rfr=RandomForestRegressor()
model_rfr.fit(x_train,y_train)
model_rfr.score(x_test,y_test)
```

```python
#xgb
from xgboost import XGBRegressor
model_xgb=XGBRegressor()
model_xgb.fit(x_train,y_train)
model_xgb.score(x_test,y_test)
```

```python
from sklearn.metrics import mean_squared_error
def predict_data(model,x_test,y_test):
    y_pred=model.predict(x_test)
    mse=mean_squared_error(y_test,y_pred)
    return print("{}".format(model),mse)
```

```python
mse_rfr=predict_data(model_rfr,x_test,y_test)
mse_rfr
```

```python
mse_xgb=predict_data(model_xgb,x_test,y_test)
mse_xgb
```

```python
#模型融合
def Mean_method(test_pred1,test_pred2):
    Mean_result=pd.concat([pd.Series(test_pred1),pd.Series(test_pred2)],axis=1).mean(axis=1)
    return Mean_result
```

```python
test_pred1=model_rfr.predict(x_test)
test_pred2=model_xgb.predict(x_test)
```

```python
mean_pred=Mean_method(test_pred1,test_pred2)
mse_mean=mean_squared_error(y_test,mean_pred)
mse_mean
```
